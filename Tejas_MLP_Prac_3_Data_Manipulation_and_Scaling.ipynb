```python
import pandas as pd

# Data
data = {
    "Car": ["Toyota", "Mitsubishi", "Skoda", "Fiat", "Mini", "VW", "Skoda", "Mercedes", "Ford", "Audi"],
    "Model": ["Aygo", "Space Star", "Citigo", "500", "Cooper", "Up!", "Fabia", "A-Class", "Fiesta", "A1"],
    "Volume": [1.0, 1.2, 1.0, 0.9, 1.5, 1.0, 1.4, 1.5, 1.5, 1.6],
    "Weight": [790, 1160, 929, 865, 1140, 929, 1109, 1365, 1112, 1150],
    "CO2": [99, 95, 95, 90, 105, 105, 90, 92, 98, 99]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV file without index
df.to_csv('cars.csv', index=False)

print("CSV file 'cars.csv' created!")
```

```python
df.head()
```

```python
#Feature Extraction
df['CO2 per CC'] = df['CO2']/(df['Volume'] * 1000)
df.head()
```

```python
#Scaler
from sklearn.preprocessing import MinMaxScaler

df2 = df.copy()
scaler = MinMaxScaler()
for i in df.columns:
  if df2[i].dtype == int or df2[i].dtype == float:
    df2[i] = scaler.fit_transform(df2[i].values.reshape(-1, 1))
df2
```

```python
from sklearn.preprocessing import StandardScaler

df3 = df.copy()
scaler = StandardScaler()
for i in df.columns:
  if df3[i].dtype == int or df3[i].dtype == float:
    df3[i] = scaler.fit_transform(df3[i].values.reshape(-1, 1))
df3
```
