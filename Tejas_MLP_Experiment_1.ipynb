```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Iris dataset, small yet enough for
df2 = pd.read_csv("/content/Iris.csv")
df2.head()
```

```python
df2.shape
```

```python
df2.columns
```

### Scatter Plot
Checking collinearity of features can be done using scatterplot.

```python
#Scatterplot to check whether Sepal Length and Sepal Width are collinear for different species
sns.scatterplot(df2, x = 'SepalLengthCm', y ='SepalWidthCm', hue = 'Species' )
plt.title("Checking Collinearity of SepalLength and SepalWidth")
```

### Histogram

```python
# Here we can use histogram to see the distribution of Sepal Length species-wise
sns.histplot(data = df2, x = "SepalLengthCm", hue = "Species", kde = True, multiple = "dodge",bins = [i for i in range(2,11)])
plt.xlim([3,10])
plt.title("Distribution of Sepal Length")
plt.show()
```

### Barplot

```python
grpd_df2 = df2.groupby("Species")["SepalLengthCm"]
grpd_df2
```

```python
ax = sns.barplot( x = grpd_df2.mean().index,y =  grpd_df2.mean().values, palette = 'pastel' )
for p in ax.patches:
  height = p.get_height() #getting height of each bar
  ax.text(p.get_x() + p.get_width()/2, height+ 0.02*height , f"{height:.2f}", ha = 'center', va = 'center' ) # X + width, height + 0.02*height --> X and Y for textual placement.
plt.ylabel("Mean Sepal Length")
plt.xlabel("Species")
plt.title("Comparison of Mean Sepal Length vs Classes")
plt.show()
```

### Pie Chart

```python
#Checking %wise distribution of the classes in iris dataset/
plt.pie(grpd_df2.count(), labels = grpd_df2.count().index,autopct = "%04.1f%%")
plt.title("%wise distribution of Species in Iris dataset")
plt.show()

```

### BoxPlot
Here we can see the median and iqr of SepalLength of each Species and outliers if any.

```python
sns.boxplot(data = df2, x ="Species",y = "SepalLengthCm", hue = "Species")
```

### Line Chart

```python
#Sepal Length across samples
plt.plot([i for i in range(len(df2))],df2.SepalLengthCm)
plt.xlabel("Sample Number")
plt.ylabel("Sepal Length")
plt.title("Sepal Length across samples")
```
