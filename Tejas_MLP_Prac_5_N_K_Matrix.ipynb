```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### N*K Matrix to show Transformations
In this practical, we show linear transformations(scaling,shrinking,shearing,reflection) showing us the linear transformations caused when multiplying a vector by a n*k matrix(n rows, k columns) here we have taken n and k to be 2.

```python
v = np.array([1,1])

transformations = {
    "Scaling": np.array([[2, 0],
                         [0, 2]]),
    "Shrinking": np.array([[0.5, 0],
                           [0, 0.5]]),
    "Shearing": np.array([[1, 1],
                          [0, 1]]),
    "Reflection": np.array([[-1, 0],
                            [0, 1]])
}
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,8))

for ax, (name, A) in zip(axes.flatten(), transformations.items()):
    v_trans = A @ v

    #Quiver used to make arrows
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original')
    ax.quiver(0, 0, v_trans[0], v_trans[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed')

    # Setting width so arrows are visible
    all_x = [v[0], v_trans[0]]
    all_y = [v[1], v_trans[1]]
    ax.set_xlim(min(all_x)-1, max(all_x)+1)
    ax.set_ylim(min(all_y)-1, max(all_y)+1)

    ax.set_aspect('equal') #X and Y are same ranged
    ax.grid(True)
    ax.set_title(name)
    ax.legend()

plt.tight_layout()
plt.show()
```
