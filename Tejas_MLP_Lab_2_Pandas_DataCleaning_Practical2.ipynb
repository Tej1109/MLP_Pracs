### Cleaning Dataset

#### Cleaning NAs

Let's start by checking whether there are any nulls in the dataset

```python
df.isna().sum()
df[df.loc[:,"Duration"] >= 120]
```

```python
## One way is to remove all rows that contan any nas
df_droppedna = df.dropna().copy() # Putting inplace = true makes it delete inplace instead of returning a dataframe,using copy cuz it returns a view of original dataframe
df_droppedna
```

```python
df_droppedna.isna().sum()
```

As we can see, no more nulls


#### Imputing Values
Another thing that can be done is replacing the nulls with a specific value or statistic like mean,median or mode.

```python
## Checking what values are na
df[df.isna().any(axis =1)]
```

```python
## Lets impute date nulls by todays date ie 27-08-2025

df.fillna({"Date" :"2025/08/27"},inplace = True)
```

```python
## Another way of imputing is by mean median or mode, Calories column has some nulls, lets impute them by mean
df.fillna({"Calories":df['Calories'].mean()},inplace = True) #We can use .median() and .mode() here too
```

```python
df.iloc[[18,28]] #As we can see the 18 and 28 rows have been imputed with mean of calories
```

#### Replacing wrong value/outliers

Say if you have identified some outliers or you have the range after which a value becomes outlier, you can fix it by manually replacing it with a value or removing it completely

```python
# I am here checking for Duration column
#Normal range of Duration is about 45 to 60, thus taking 120 as upper limit lets check how many rows are above it

df[df.Duration >= 120]
```

```python
#Lets fix Duration above 120 to be 120
for x in df.index:
  if df.loc[x,"Duration"] >= 120:
    df.loc[x,"Duration"] = 120
```

```python
#Lets check if anything stil present
for x in df.index:
  if df.loc[x,"Duration"] > 120:
    print(df.loc[x])
#Nothing is present!

df[df.Duration == 120] #Instead that old value has been changed to 120
```

```python
#Another thing that u can do is drop the outlier value
df_droppedna[df_droppedna["Duration"] > 120]
```

```python
for x in df_droppedna.index:
  df_droppedna.drop(x,inplace = True)
df_droppedna[df_droppedna["Duration"] > 120] #No more values available
```
